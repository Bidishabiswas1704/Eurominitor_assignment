# I. Ice Cream Dataset

## Data Loading and Exploration and Preprocessing

```python
import pandas as pd

# Load the CSV data into a DataFrame
ice_cream_data = pd.read_csv("ice.cream.data.csv")

# Define the range of years from 2000 to 2019
years = [str(year) for year in range(2000, 2019 + 1)]

# Step 1: Filter to keep only Retail Volume in tonnes and Retail Value in USD
# Ensure 'Data Type' exists before subsetting
if 'Data Type' not in ice_cream_data.columns:
    raise KeyError("'Data Type' column is missing in the original data.")

ice_cream_data = ice_cream_data[
    ice_cream_data["Data Type"].isin(["Retail Volume (Tonnes)", "Retail Value RSP"])
]

# Step 2: Keep essential columns and remove unnecessary ones
required_columns = ["Region", "Country", "Data Type", "Unit", "Unit Multiplier"] + years
ice_cream_data = ice_cream_data[required_columns]

# Step 3: Convert volume to kilograms and value to US dollars
# Define unit multipliers for conversion
unit_multipliers = {
    "000 tonnes": 1_000,  # Conversion for volume
    "USD million": 1_000_000  # Conversion for value
}

# Apply the conversions
for year in years:
    # Convert Retail Volume to kilograms
    retail_volume_mask = (ice_cream_data["Data Type"] == "Retail Volume (Tonnes)")
    ice_cream_data.loc[retail_volume_mask, year] *= unit_multipliers["000 tonnes"]
    
    # Convert Retail Value RSP to US dollars
    retail_value_mask = (ice_cream_data["Data Type"] == "Retail Value RSP")
    ice_cream_data.loc[retail_value_mask, year] *= unit_multipliers["USD million"]

# Step 4: Remove "Unit" and "Unit Multiplier" after conversion
ice_cream_data.drop(columns=["Unit", "Unit Multiplier"], inplace=True)

# Step 5: Reshape the Data with pd.melt() to Keep It Simple
ice_cream_data_long = pd.melt(
    ice_cream_data,
    id_vars=["Region", "Country", "Data Type"],
    value_vars=years,
    var_name="Year",
    value_name="Value"
)

# Ensure 'Data Type' exists before further operations
if 'Data Type' not in ice_cream_data_long.columns:
    raise KeyError("'Data Type' column is missing before proceeding.")

# Display the reshaped DataFrame to verify the output
print("Reshaped DataFrame:")
ice_cream_data_long.head(30)
```

### Convert data to the analysis friendly format

```python
# Pivot the data to convert it to the desired format
reshaped_data_wide = ice_cream_data_long.pivot_table(
    index=["Region", "Country", "Year"],
    columns="Data Type",
    values="Value"
).reset_index()

# Display the resulting DataFrame
print("Reformatted DataFrame:")
print(reshaped_data_wide)
```

### Rename variables to “Ice Cream Value” and “Ice Cream Volume”

```python
# Renaming the variables
reshaped_data_wide.rename(
    columns={
        "Retail Volume (Tonnes)": "Ice Cream Volume",
        "Retail Value RSP": "Ice Cream Value"
    },
    inplace=True
)

# Display the DataFrame after renaming
print("After Renaming Variables:")
reshaped_data_wide.head(30)
```

### Additional column, Ice Cream Price is added

```python
# Ensure that "Ice Cream Volume" and "Ice Cream Value" columns exist
if "Ice Cream Volume" not in reshaped_data_wide.columns or "Ice Cream Value" not in reshaped_data_wide.columns:
    raise KeyError("'Ice Cream Volume' or 'Ice Cream Value' column is missing.")

# Calculate Ice Cream Price
reshaped_data_wide["Ice Cream Price"] = reshaped_data_wide["Ice Cream Value"] / reshaped_data_wide["Ice Cream Volume"]

# Display the DataFrame with Ice Cream Price
print("With Ice Cream Price:")
reshaped_data_wide.head(30)
```

### Each years, total volume of Icecream by Region

```python
# Calculate total volume by Region for each Year
total_volume_by_region = reshaped_data_wide.groupby(["Year", "Region"])["Ice Cream Volume"].sum()

# Display the total volume by Region for each Year
print("Total Volume by Region for Each Year:")
total_volume_by_region.tail(20)
```

```python
# Calculate total volume by Region for each Year
total_volume_by_region = reshaped_data_wide.groupby(["Year", "Region"])["Ice Cream Volume"].sum()

# Reset the index to make Year and Region columns accessible
total_volume_by_region_reset = total_volume_by_region.reset_index()

# Display the structure of the total_volume_by_region_reset dataframe
print(total_volume_by_region_reset.info())

# Check the unique values in the "Year" column
print(total_volume_by_region_reset['Year'].unique())
```

### Finding second largest region by ice cream volumes in 2018

```python
# Convert the "Year" column to integer type
total_volume_by_region_reset['Year'] = total_volume_by_region_reset['Year'].astype(int)

# Filter the data for the year 2018
volume_2018 = total_volume_by_region_reset[total_volume_by_region_reset['Year'] == 2018]

# Sort the regions by total ice cream volumes in descending order
sorted_regions_2018 = volume_2018.sort_values(by='Ice Cream Volume', ascending=False)

# Display the sorted regions for 2018
print("Sorted Regions for 2018:")
print(sorted_regions_2018)

```

```python
# Sort the DataFrame in descending order by Ice Cream Volume
sorted_regions_2018 = volume_2018.sort_values(by="Ice Cream Volume", ascending=False)

# Select the second row from the sorted DataFrame (indexing starts from 0)
second_largest_region = sorted_regions_2018.iloc[1]

# Display the second largest region by ice cream volumes in 2018
print("Second largest region by ice cream volumes in 2018:")
print(second_largest_region)
```

### Finding:
#### a)	What is the third largest Region by Ice Cream Value?

#### b)	What is the tenth largest country by Ice Cream Volume?

#### c)	Which country had the second largest Ice Cream Price in 2009?


```python
def calculate_total_volume_by_region(data):
    """
    Calculate total volume by region for each year.
    
    Parameters:
    - data: DataFrame containing the data
    
    Returns:
    - DataFrame with total volume by region for each year
    """
    total_volume_by_region = data.groupby(["Year", "Region"])["Ice Cream Volume"].sum().reset_index()
    return total_volume_by_region
```

```python
# Use the function to calculate total volume by region for each year
total_volume_by_region = calculate_total_volume_by_region(reshaped_data_wide)

# a) What is the third largest Region by Ice Cream Value?
third_largest_region = total_volume_by_region.groupby('Region')['Ice Cream Volume'].sum().nlargest(3).index[-1]
print("Third largest Region by Ice Cream Value:", third_largest_region)

# b) What is the tenth largest country by Ice Cream Volume?
tenth_largest_country = total_volume_by_region.groupby('Region')['Ice Cream Volume'].sum().nlargest(10).index[-1]
print("Tenth largest Region by Ice Cream Volume:", tenth_largest_country)

# c) Which country had the second largest Ice Cream Price in 2009?
second_largest_price_country_2009 = reshaped_data_wide[reshaped_data_wide['Year'] == '2009'].nlargest(2, 'Ice Cream Price')['Country'].iloc[1]
print("Country with the second largest Ice Cream Price in 2009:", second_largest_price_country_2009)
```

# II. Macro Dataset

## Data Loading and Exploration

```python
import pandas as pd

# Load the CSV data into a DataFrame
macro_data = pd.read_csv("macro.data.csv")


# Display the column names
print("Column Names:")
print(macro_data.columns)
```

## Data Preprocessing

```python
# Define the range of years
years = [str(year) for year in range(1977, 2030 + 1)]

# Step 1: Keep essential columns and remove unnecessary ones
required_columns = ["Region", "Country", "Subcategory", "Unit", "Unit Multiplier"] + years
macro_data = macro_data[required_columns]

# Step 2: Define unit multipliers for conversion
unit_multipliers = {
    "USD million": 1000000,
    "% of households": 1,
    "0": 1000
}

# Step 3: Apply the conversions
for year in years:
    # Select the data for the current year
    selected_data = macro_data[["Subcategory", "Unit", "Unit Multiplier", year]]
    
    # Convert the values based on the unit multiplier
    selected_data[year] *= selected_data["Unit Multiplier"].map(unit_multipliers)
    
    # Drop the 'Unit' and 'Unit Multiplier' columns as they are no longer needed
    selected_data.drop(columns=["Unit", "Unit Multiplier"], inplace=True)

# Step 4: Reshape the Data with pd.melt() to Keep It Simple
macro_data_long = pd.melt(
    macro_data,
    id_vars=["Region", "Country", "Subcategory"],
    value_vars=years,
    var_name="Year",
    value_name="Value"
)

# Ensure 'Subcategory' exists before further operations
if 'Subcategory' not in macro_data_long.columns:
    raise KeyError("'Subcategory' column is missing before proceeding.")

# Display the reshaped DataFrame to verify the output
print("Reshaped DataFrame:")
macro_data_long.head(30)
```

```python
# Pivot the data to convert it to the desired format
reshaped_micro_wide = macro_data_long.pivot_table(
    index=["Region", "Country", "Year"],
    columns="Subcategory",
    values="Value"
).reset_index()

# Display the resulting DataFrame
print("Reformatted DataFrame:")
reshaped_micro_wide.head(30)
```

```python
# Convert 'Year' column to integers
reshaped_micro_wide['Year'] = reshaped_micro_wide['Year'].astype(int)

# Filter micro data to include only the years from 2000 to 2020
micro_data_common_years = reshaped_micro_wide[(reshaped_micro_wide['Year'] >= 2000) & (reshaped_micro_wide['Year'] <= 2020)]

# Display the filtered micro data
print("Filtered Micro Data:")
micro_data_common_years.head(30)
```

# III. Data Merging

```python
# Get information about reshaped_data_wide
print("Information about reshaped_data_wide:")
print(reshaped_data_wide.info())

# Get information about micro_data_common_years
print("\nInformation about micro_data_common_years:")
print(micro_data_common_years.info())
```

```python
# Convert 'Year' column in reshaped_data_wide to int32
reshaped_data_wide['Year'] = reshaped_data_wide['Year'].astype(int)

# Merge reshaped_data_wide with micro_data_common_years
merged_data = pd.merge(reshaped_data_wide, micro_data_common_years, on=["Region", "Country", "Year"], how="inner")

# Display the merged DataFrame
print("Merged DataFrame:")
merged_data.head(30)
```

## Calculating Ice Cream Volume per Capita and GDP per Capita

```python
# Remove leading spaces from column names
merged_data.columns = merged_data.columns.str.strip()

# Calculate per capita variables
merged_data['GDP per Capita'] = merged_data['Total GDP'] / merged_data['Population: National Estimates at January 1st']
merged_data['Ice Cream Volume per Capita'] = merged_data['Ice Cream Volume'] / merged_data['Population: National Estimates at January 1st']
```

```python
# Display the first few rows of the merged dataset
merged_data.head()
```

## Visualizing the relationship between Ice Cream Volume per Capita and GDP per Capita 

```python
import os
import matplotlib.pyplot as plt

# Create the Output directory if it doesn't exist
output_dir = 'Output'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Plot Ice Cream Volume per Capita vs GDP per Capita
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['GDP per Capita'], merged_data['Ice Cream Volume per Capita'], alpha=0.5)
plt.title('Ice Cream Volume per Capita vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Ice Cream Volume per Capita')
plt.grid(True)

# Save the plot to the Output folder
plt.savefig(os.path.join(output_dir, 'Ice_Cream_vs_GDP_Per_Capita.png'))

# Show the plot
plt.show()
```

## Calculating Missing data percentage

```python
import pandas as pd
import os

def calculate_missing_data_percent(data):
    # Calculate the percentage of missing data for each column
    missing_data_percent = (data.isnull().sum() / len(data)) * 100
    
    return missing_data_percent

# Calculate missing data percentage for reshaped_micro_wide DataFrame
missing_data_percent_reshaped_micro_wide = calculate_missing_data_percent(reshaped_micro_wide)

# Calculate missing data percentage for reshaped_data_wide DataFrame
missing_data_percent_reshaped_data_wide = calculate_missing_data_percent(reshaped_data_wide)

# Create the Output directory if it doesn't exist
output_dir = 'Output'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Write the table to CSV files in the Output folder
output_file_path_reshaped_micro_wide = os.path.join(output_dir, 'missing_data_percent_reshaped_micro_wide.csv')
missing_data_percent_reshaped_micro_wide.to_csv(output_file_path_reshaped_micro_wide, header=True)

output_file_path_reshaped_data_wide = os.path.join(output_dir, 'missing_data_percent_reshaped_data_wide.csv')
missing_data_percent_reshaped_data_wide.to_csv(output_file_path_reshaped_data_wide, header=True)

# Display the missing data percentage for reshaped_micro_wide DataFrame
print("Missing Data Percentage for reshaped_micro_wide DataFrame:")
print(missing_data_percent_reshaped_micro_wide)

# Display the missing data percentage for reshaped_data_wide DataFrame
print("\nMissing Data Percentage for reshaped_data_wide DataFrame:")
print(missing_data_percent_reshaped_data_wide)
```
